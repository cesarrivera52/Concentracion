## **Filtros basicos para variables cuantitativas (variables numericas)**

```python
#Filtro comparacion
filtro6=df[df['MONTO ADEUDADO']>70_000]
filtro6
```

## **Filtros basicos con conectores logicos**

```python
#Filtro &
filtro7=df[(df['MONTO ADEUDADO']<6_000)&(df['FECHA_FACTURA']=='2023-03-14')]
filtro7
```

```python
#Filtro OR
filtro8=df[(df['MONTO ADEUDADO']<6_000)|(df['FECHA_FACTURA']=='2023-03-14')]
filtro8
```

```python
#Filtro NOT
filtro9=df[~(df['MONTO ADEUDADO']<6_000)& ~(df['FECHA_FACTURA']=='2023-03-14')]
filtro9
```

```python
filtro9.to_csv('Entregable1.csv')
```
