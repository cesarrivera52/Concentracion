```python
#Importamos librerias requeridas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Utilizar el nuevo repositorio con el nombre: Actividad 1

Agregar el archivo: Inmuebles.xlsx

Aplicar los siguientes filtros:





**a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”**

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Inmuebles.xlsx',skiprows=1)
```

```python
df.head()
```

```python
#Verificamos informacion del DataFrame
df.info()
```

```python
filtroa=df.drop('Referencia',axis=1).head()
filtroa.head()
```

```python
filtroa.to_csv('Filtro_A_Act3.1.3.csv')
```


b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00



```python
filtrob=df[(df['Fecha_Alta']=='2004-01-02')|(df['Fecha_Alta']=='2007-04-29')]
filtrob.head()
```

```python
filtrob.to_csv('Filtro_B_Act3.1.3.csv')
```

c) Todos los registros de “ Tipo: parking, local, industrial y piso”



```python
filtroc=df[(df['Tipo']=='Parking')|(df['Tipo']=='Local')|(df['Tipo']=='Industrial')|(df['Tipo']=='Piso')]
filtroc.head()
```

```python
filtroc.to_csv('Filtro_C_Act3.1.3.csv')
```

d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”


```python
filtrod = df[(df['Provincia'] == 'Girona') & (df['Superficie'] > 100)]
filtrod.head()
```

```python
filtrod.to_csv('Filtro_D_Act3.1.3.csv')
```

e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”


```python
filtroe = df[(df['Precio Venta'] > 1_000_000) | (df['Vendedor']=='Carmen')|(df['Vendedor']=='Joaquin')]
filtroe.head()
```

```python
filtroe.to_csv('Filtro_E_Act3.1.3.csv')
```

f) Todos los registros que no sean de “Provincia : Barcelona“


```python
filtrof = df[(df['Provincia'] != 'Barcelona')]
filtrof.head()
```

```python
filtrof.to_csv('Filtro_F_Act3.1.3.csv')
```

g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“

```python
filtrog = df[(df['Operación'] != 'Alquiler')&(df['Tipo'] != 'Oficina')]
filtrog.head()
```

```python
filtrog.to_csv('Filtro_G_Act3.1.3.csv')
```
