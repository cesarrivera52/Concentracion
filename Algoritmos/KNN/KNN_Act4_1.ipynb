# Paso 0 **Importar librerias y datos**

```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
```

```python
data= {'petallength': [3.5,5.8],
       'petalwidth': [1,1.6],
       'class': ['Iris-versicolor','Iris-virginica']}

clasificacion= pd.DataFrame(data)
clasificacion
```

```python
entrenamiento=pd.read_csv('Entrenamiento.csv')
entrenamiento.head()
```

# Paso 1 **Determinar clases**
Separamos las clases de los tipos de flores para poder determinar las clases

```python
versicolor= entrenamiento[entrenamiento['class']=='Iris-versicolor']
virginica= entrenamiento[entrenamiento['class']=='Iris-virginica']
```

# Paso 2 **Graficar las clases**

Se grafican las clases para poder observar donde se ubican los puntos

```python
plt.scatter(versicolor['petallength'],versicolor['petalwidth'],
            marker='*',s=150, color='blue',
            label='Versicolor (Clase 1)')

plt.scatter(virginica['petallength'],virginica['petalwidth'],
            marker='.',s=150, color='red',
            label='Virginica (Clase 2)')

plt.ylabel("Petal Width")
plt.xlabel('Petal Lenght')
plt.legend(bbox_to_anchor=(1,0.2))
plt.show
```

# Paso 3 **Creación del Modelo KNN**
Valor de K= El valor de k va a ser el mismo para ambas instancias

# Instancia 0


*   Petal lenght: 3.5
*   Petal width: 1.0





```python
df_Nuevo=entrenamiento
```

```python
clasificacion
```

```python
prueba_x_petallength=(3.5)
prueba_y_petalwidth=(1)
```

```python
plt.scatter(versicolor['petallength'],versicolor['petalwidth'],
            marker='*',s=150, color='skyblue',
            label='Versicolor (Clase 1)')

plt.scatter(virginica['petallength'],virginica['petalwidth'],
            marker='.',s=150, color='red',
            label='Virginica (Clase 2)')

plt.scatter(prueba_x_petallength,prueba_y_petalwidth,
            marker='.',s=150, color='purple',
            label='Clase ?')

plt.ylabel("Petal Width")
plt.xlabel('Petal Lenght')
plt.legend(bbox_to_anchor=(1,0.2))
plt.show
```

EUCLIDEAN DIST

```python
df_Nuevo['Euclidean Distance']=np.sqrt((np.power(entrenamiento['petallength']-prueba_x_petallength,2)) + (np.power(entrenamiento['petalwidth']-prueba_y_petalwidth,2)))

df_Nuevo['Manhattan Distance']= (np.abs(entrenamiento['petallength']-prueba_x_petallength))+(np.abs(entrenamiento['petalwidth']-prueba_y_petalwidth))
```

```python
df_Nuevo.head()
```

```python
vecinos=int(input('Ingrese el numero de vecinos:  '))
```

```python
df_eucledian=df_Nuevo.sort_values(by='Euclidean Distance').head(vecinos)
```

MANHATTAN DIST


```python
df_manhattan=df_Nuevo.sort_values(by='Manhattan Distance').head(vecinos)
```

Si es por euclediana

```python
Tmp=df_eucledian['class'].value_counts()

print('Predicción con dist. euclidiana: ',Tmp[Tmp==Tmp.max()].index.values[0])
print('Clasificación Real:  ',clasificacion['class'][0])
```

Si es por manhattan

```python
Tmp_man=df_manhattan['class'].value_counts()

print('Predicción con dist. Manhattan: ',Tmp_man[Tmp_man==Tmp_man.max()].index.values[0])
print('Clasificación Real:  ',clasificacion['class'][0])
```
