```python
import pandas as pd
```

```python
data=pd.read_csv('listings.csv')
data.info()
```

```python
data.drop(['listing_url','scrape_id','last_scraped','source','name','description','bathrooms_text','neighborhood_overview','picture_url','host_url','host_about','host_thumbnail_url','host_picture_url','neighbourhood_group_cleansed','bathrooms','amenities','minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights','calendar_updated','calendar_last_scraped','number_of_reviews_l30d','first_review','license','calculated_host_listings_count_entire_homes','last_review','calculated_host_listings_count_private_rooms','calculated_host_listings_count_shared_rooms'],axis=1,inplace=True)
```

```python
data.info()
```

```python
nulos=data.isnull().sum().reset_index()
nulos=nulos[nulos[0]!=0]
nulos
```

```python
for i in nulos['index']:
    try:
        float(nulos[i][0])
        data[i]=data[i].fillna(data[i].mean())
    except:
        data[i]=data[i].fillna(method='ffill')
```

```python
nulos=data.isnull().sum().reset_index()
nulos=nulos[nulos[0]!=0]
nulos
```

```python
data.dropna(inplace=True)
```

```python
nulos=data.isnull().sum().reset_index()
nulos=nulos[nulos[0]!=0]
nulos
```

```python
cualitativos=pd.DataFrame()
cuantitativos=pd.DataFrame()
```

```python
data.info()
```

```python
iterador1=data.dtypes.reset_index()
iterador=iterador1[0]
iterador1=iterador1[['index']]
```

```python
iterador1
```

```python
iterador
```

```python
contador=0
for i in iterador:
    print(i)
    if i=='object':
        cualitativos=pd.concat([cualitativos,data[data.columns[contador]]],axis=1)
    else:
        cuantitativos=pd.concat([cuantitativos,data[data.columns[contador]]],axis=1)
    contador+=1
```

```python
cualitativos.to_csv('cualitativos.csv')
```

```python
cuantitativos.to_csv('cuantitativos.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
#Realizamos diagrama de caja o bigote de cada columna del DF
fig= plt.figure(figsize=(15 ,8))
cuantitativos['reviews_per_month'].plot(kind='box', vert=False)
plt.title('Valores Atípicos del Malaga')
plt.show() #Dibujamos el diagrama
```

```python
y=cuantitativos
Limite_Superior= y.mean()+ 3*y.std()
Limite_Inferior= y.mean()- 3*y.std()
print('-Límite superior permitido\n', Limite_Superior,'\n')
print('-Límite inferior permitido\n', Limite_Inferior)
```

```python
data2= cuantitativos[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
valores_nulos= data2.isnull().sum()
valores_nulos
```

```python
data_clean=data2.copy()
data_clean=data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos= data_clean.isnull().sum()
valores_nulos
```

```python
y=cuantitativos

percentiles25=y.quantile(0.25) #Q1
percentiles75=y.quantile(0.75) #Q3
iqr= percentiles75-percentiles25

Limite_Superior_iqr=percentiles75 + 1.5*iqr
Limite_Inferior_iqr=percentiles25- 1.5*iqr
print("Límite superior permitido: \n",Limite_Superior_iqr)
print("\nLímite inferior permitido: \n",Limite_Inferior_iqr)
```

```python
data2_iqr=cuantitativos[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data2_iqr
```

```python
valores_nulos=data2_iqr.isnull().sum()
valores_nulos
```

```python
data_3_iqr= data2_iqr.copy()
data_3_iqr=data_3_iqr.fillna(round(data2_iqr.mean(),1))
```

```python
fig= plt.figure(figsize=(15 ,8))
data_clean['reviews_per_month'].plot(kind='box', vert=False)
plt.title('Valores Atípicos Procedimiento "Desviación Estandar" para eliminar outliers en el DF')
plt.show() #Dibujamos el diagrama
```

```python
fig = plt.figure(figsize =(15, 8))
data_3_iqr['reviews_per_month'].plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
```
