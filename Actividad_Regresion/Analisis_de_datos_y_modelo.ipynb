```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dependientes= pd.read_csv('dependientes.csv')
dependientes.drop(columns=['Unnamed: 0'],inplace=True)
dependientes.head()
```

```python
dependientes.info()
```

```python
independientes= pd.read_csv('independientes.csv')
independientes.drop(columns=['Unnamed: 0'],inplace=True)
independientes.head()
```

```python
independientes.info()
```

```python
def graficar(y,x):
  plt.figure()
  plt.scatter(x['host_response_rate'],y['host_acceptance_rate'])
  plt.xlabel('host_response_rate')
  plt.ylabel('host_acceptance_rate')
  plt.title('host_acceptance_rate vs host_response_rate')
  plt.show()
  plt.figure()
  plt.scatter(x['review_scores_cleanliness'],y['review_scores_location'])
  plt.xlabel('review_scores_cleanliness')
  plt.ylabel('review_scores_location')
  plt.title('review_scores_location vs review_scores_cleanliness')
  plt.show()
  plt.figure()
  plt.scatter(x['price'],y['host_acceptance_rate'])
  plt.xlabel('price')
  plt.ylabel('host_acceptance_rate')
  plt.title('price vs host_acceptance_rate')
  plt.show()
  plt.figure()
  plt.scatter(x['number_of_reviews'],y['availability_365'])
  plt.xlabel('number_of_reviews')
  plt.ylabel('availability_365')
  plt.title('availability_365 vs number_of_reviews')
  plt.show()
  plt.figure()
  plt.scatter(x['number_of_reviews'],y['host_acceptance_rate'])
  plt.xlabel('number_of_reviews')
  plt.ylabel('host_acceptance_rate')
  plt.title('host_acceptance_rate vs number_of_reviews')
  plt.show()
  plt.figure()
  plt.scatter(x['review_scores_communication'],y['reviews_per_month'])
  plt.xlabel('review_scores_communication')
  plt.ylabel('reviews_per_month')
  plt.title('reviews_per_month vs review_scores_communication')
  plt.show()
```

```python
graficar(dependientes[dependientes['clave_ciudad']=='CDMX'],independientes[independientes['clave_ciudad']=='CDMX'])
```

```python
graficar(dependientes[dependientes['clave_ciudad']=='MALAGA'],independientes[independientes['clave_ciudad']=='MALAGA'])
```

```python
graficar(dependientes[dependientes['clave_ciudad']=='RIO'],independientes[independientes['clave_ciudad']=='RIO'])
```
