## **Filtros basicos para variables cuantitativas (variables numericas)**

```python
#Filtro comparacion
filtro6=df[df['PRECIO_UNITARIO']<100]
filtro6
```

```python
filtro6.to_csv('Filtro6.csv')
```

## **Filtros basicos con conectores logicos**

```python
#Filtro &
filtro7=df[(df['CANT']>4)&(df['PRECIO_UNITARIO']<30)]
filtro7
```

```python
filtro7.to_csv('Filtro7.csv')
```

```python
#Filtro OR
filtro8=df[(df['CANT']==9)|(df['PRECIO_UNITARIO']<50)]
filtro8
```

```python
filtro8.to_csv('Filtro8.csv')
```

```python
#Filtro NOT
filtro9=df[~(df['CANT']==9)&~(df['PRECIO_UNITARIO']<50)]
filtro9
```

```python
filtro9.to_csv('Filtro9.csv')
```
