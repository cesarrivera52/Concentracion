```python
def strugens(city,columna):
    import numpy as np
    #n=city.shape[0]
    Max=city[columna].max()
    Min=city[columna].min()
    R=Max-Min
    ni= 1+3.32*np.log10(Max)
    #i=R/ni
    return np.linspace(Min,Max,round(ni))
```
